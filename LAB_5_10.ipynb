```python
# Gerar um gráfico de barras com o numpy
import numpy as np

x = np.arange(50)
y = np.random.rand(50)

plt.bar (x, y, color = 'pink')
plt.xlabel ("Eixo X")
plt.ylabel ("Eixo Y")
plt.legend ()
plt.title ("Gráfico de barras criado com numpy")
plt.show()
```

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,5,6,8,4,8]

'''
Usando o marker:
 'o' = bolinha
 '|' = |
 'v' = triangulo
 'x' = x
'''

plt.scatter(x,y, label = 'Pontos', color = 'r', marker = 'v', s = 80)
plt.legend ()
plt.show()
```

Stack Plots

```python
dias = [1,2,3,4,5]
dormir = [7,8,6,77,7]
comer = [2,3,4,5,3]
trabalhar = [7,8,7,2,2]
```

```python
plt.stackplot (dias,dormir,comer,trabalhar,
               labels = ['dias', 'dormir', 'comer', 'trabalhar'],
               colors = ['m','c','r','k'])
plt.legend(loc='upper left')
plt.show()


```
